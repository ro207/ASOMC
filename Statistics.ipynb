```python
import pandas as pd
```

```python
stats = pd.read_csv('Stats.csv').dropna()
stats
```

```python
#Filtering
successes = stats[stats['outcome'] == 'Survived']
successes
```

```python
#Sorting
stats[stats['outcome'] == 'defeated'][['time_survived','score']].plot(kind = 'scatter', x = 'time_survived', y = 'score', xlabel = 'Time Survived (ms)', ylabel = 'Score', title = 'Time Survived vs. Score')
```

Therefore, there is a strong positive correlation between the time survived and the score.

```python
stats[['health_lost','fireballs_hit']].plot(kind = 'scatter', x = 'fireballs_hit', y = 'health_lost', xlabel = 'Fireballs Hit', ylabel = 'Health Lost', title = 'Fireballs Hit vs. Health Lost')
```

Therefore, there is little to no positive correlation between fireballs hit and health lost.
